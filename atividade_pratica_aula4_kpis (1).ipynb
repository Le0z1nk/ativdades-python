## 7. Análise por canal

### Pergunta de negócio
**Qual canal de venda é o mais eficiente?**

### Tarefa
Crie uma tabela agregada por `canal_venda` contendo, no mínimo:
- Receita Total
- Lucro Total
- Margem Média ou Margem Calculada
- Quantidade de pedidos
- Ticket Médio

### Depois responda
1. Qual canal gera mais receita?
2. Qual canal gera maior eficiência?
3. Se você fosse gestor, priorizaria volume ou margem?


```python
canal_venda_analise = df.groupby('canal_venda').agg(
    receita_total=('receita', 'sum'),
    lucro_total=('lucro', 'sum'),
    quantidade_pedidos=('pedido_id', 'count')
).reset_index()

canal_venda_analise['margem_calculada'] = canal_venda_analise['lucro_total'] / canal_venda_analise['receita_total'] *100
canal_venda_analise['ticket_medio'] = canal_venda_analise['receita_total'] / canal_venda_analise['quantidade_pedidos']

display(canal_venda_analise.sort_values(by='receita_total', ascending=False))

# O canal que gera mais receita é o marketplace
# O canal que gera maior eficiencia é o online
# Se eu fosse gestor priorizaria margem
```

### Insight obrigatório
Escreva 1 ou 2 frases interpretando sua tabela, no estilo executivo:
- não apenas “o Online foi maior”

De acordo com os dados, o canal Marketplacle foi o que gerou a amior receita, mas o canal Online tem a maior margem de lucro. Por isso, o foco atual deve ser voltado ao canal de vendas online.

## 8. Análise por categoria

### Pergunta de negócio
**Qual categoria é mais lucrativa e qual parece mais eficiente?**

### Tarefa
Crie uma tabela agregada por `categoria` com:
- Receita Total
- Lucro Total
- Margem
- Ticket Médio

### Depois responda
1. A categoria de maior receita também é a de maior margem?
2. Existe alguma categoria que parece vender bem, mas ser pouco eficiente?


```python
categoria_analise = df.groupby('categoria').agg(
    receita_total=('receita', 'sum'),
    lucro_total=('lucro', 'sum'),
    quantidade_pedidos=('pedido_id', 'count')
).reset_index()

categoria_analise['margem'] = categoria_analise['lucro_total'] / categoria_analise['receita_total'] * 100
categoria_analise['ticket_medio'] = categoria_analise['receita_total'] / categoria_analise['quantidade_pedidos']

display(categoria_analise.sort_values(by='receita_total', ascending=False))

# A categoria de maior receita é a de telefonia, mas a categoria de maior margem é a de periféricos
# A categoria de telefonia, apesar de vender muito bem, é pouco eficiente por ter uma margem de lucro baixa
```

## 9. Análise por região (UF)

### Pergunta de negócio
**Quais UFs concentram a operação?**

### Tarefa
Monte uma tabela por `uf` com:
- Receita Total
- Lucro Total
- Participação percentual da receita
- Ticket Médio

### Depois responda
1. Quais UFs concentram maior receita?
2. Existe concentração regional?
3. Se fosse necessário expandir ou reforçar logística, por onde você começaria?


```python
uf_analise = df.groupby('uf').agg(
    receita_total=('receita', 'sum'),
    lucro_total=('lucro', 'sum'),
    quantidade_pedidos=('pedido_id', 'count')
).reset_index()

receita_total_global = uf_analise['receita_total'].sum()
uf_analise['participacao_percentual_receita'] = (uf_analise['receita_total'] / receita_total_global) * 100
uf_analise['ticket_medio'] = uf_analise['receita_total'] / uf_analise['quantidade_pedidos']

display(uf_analise.sort_values(by='receita_total', ascending=False))

# Os UFs que concentram a maior receita são: SP, RJ e PE
# Não existe concentração regiional
# Começaria pelos UFs que tem uma participação percentual da receita menor
```

## 12. Análise temporal

A aula reforça que sem tempo a análise fica estática. fileciteturn5file0

### Tarefa
1. Converta `data_venda` para data, se necessário
2. Crie colunas auxiliares como:
   - mês
   - trimestre (opcional)
3. Gere uma tabela agregada por mês com:
   - receita
   - lucro
   - ticket médio ou margem

### Depois responda
1. O lucro parece crescer ao longo do tempo?
2. O último trimestre foi melhor que o anterior?
3. Quais meses concentram as vendas de fim de ano?


```python
data_analise = df.groupby(df['data_venda'].dt.to_period('M')).agg(

    receita_total=('receita', 'sum'),
    lucro_total=('lucro', 'sum'),
).reset_index()

data_analise['margem'] = data_analise['lucro_total'] / data_analise['receita_total']

display(data_analise.sort_values(by='data_venda', ascending=True))

# O lucro não parece crescer ao longo do tempo
# O último trimestre foi melhor que o anterior
# Os meses que concentram as vendas de fim de ano são: novembro e dezembro
```
